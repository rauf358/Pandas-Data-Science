```python
hacks["created_at"]=pd.to_datetime(hacks["created_at"])
```

```python
ask = hacks['title'].str.lower().str.startswith('ask hn')
show = hacks['title'].str.lower().str.startswith('show hn')
apply = hacks['title'].str.lower().str.startswith('apply hn')
how = hacks['title'].str.lower().str.startswith('how')
```

```python
ask_post=hacks[ask]
show_post=hacks[show]
apply_post=hacks[apply]
how_post=hacks[how]
```

```python
ask_post # created ask_post DATAFRAME
```

```python
ask_post.info()
```

```python
ask_post["hour"]=ask_post["created_at"].dt.hour # created a column in Dataframe
```

```python
ask_post
```

```python
avg_comnt_ask =ask_post.pivot_table(index='hour',values='num_comments',aggfunc='mean')
avg_comnt_ask # recorded comments on hourly basis
```

```python
top_ask_hour=avg_comnt_ask.sort_values("num_comments",ascending=False)#sorted values as par to
top_5_ask_hour=top_ask_hour.head()#getting most comments in one hour
top_5_ask_hour
```

```python
show_post["hour"]=show_post["created_at"].dt.hour 
how_post["hour"]=how_post["created_at"].dt.hour
apply_post["hour"]=apply_post["created_at"].dt.hour 
```

```python
avg_comnt_show =show_post.pivot_table(index='hour',values='num_comments')
avg_comnt_how =how_post.pivot_table(index='hour',values='num_comments')
avg_comnt_apply =apply_post.pivot_table(index='hour',values='num_comments')
```

```python
top_show_hour=avg_comnt_show.sort_values("num_comments",ascending=False)
top_5_show_hour=top_show_hour.head()
top_5_show_hour
```

```python
top_how_hour=avg_comnt_how.sort_values("num_comments",ascending=False)
top_5_how_hour=top_how_hour.head()
top_5_how_hour
```

```python
top_apply_hour=avg_comnt_apply.sort_values("num_comments",ascending=False)
top_5_apply_hour=top_apply_hour.head()
top_5_apply_hour
```
