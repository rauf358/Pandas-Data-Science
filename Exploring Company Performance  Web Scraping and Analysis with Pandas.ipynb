This project focuses on extracting data from a website using web scraping techniques and conducting analysis on the extracted data using the pandas library in Python. The goal is to gather information about various companies, including their revenue, number of employees, and revenue per employee.

```python
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_largest_technology_companies_by_revenue'
```

```python
response = requests.get(url)
```

```python
response
```

```python
text = response.text
```

```python
text
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(text,'html')
```

```python
soup
```

```python
tables = soup.find('table',{"class":"wikitable sortable plainrowheads"})
```

```python
tables
```

```python
rows = tables.find_all('tr')
```

```python
rows
```

```python
dir(tables)
```

```python
headers=[]
for row in rows:
    th = row.find_all('th') # find list of <>
    for h in th:
        headers.append(h.text.strip())
headers
```

```python
final_data=[]
for row in rows:
    td = row.find_all('td')
    data=[]
    for d in td:
        data.append(d.text.strip())
    final_data.append(data)
final_data
```

```python
final_data = final_data[1:]
```

```python
import pandas as pd
```

```python
top = pd.DataFrame(final_data)
top
```

```python
dir(top.columns)
```

```python
def con_vert(data):
    if '$' in data:
        data=data.replace("$",'')
        data=float(data)
    elif ',' in data:
        data=data.replace(",",'')
        data=int(data)
    else:
        data=float(data)
    return data
```

```python
top['Revenue ($B) USD[2]']=top['Revenue ($B) USD[2]'].apply(con_vert)
```

```python
top['Employees[2]']=top['Employees[2]'].apply(con_vert)
```

```python
top['Revenue per employee ($K USD)[2]']=top['Revenue per employee ($K USD)[2]'].apply(con_vert)
```

```python
top
```

```python
top.rename({'Revenue ($B) USD[2]':'Revenue ($B) USD','Employees[2]':'Employees','Revenue per employee ($K USD)[2]':'Revenue per employee ($K USD)'},axis=1,inplace=True)
```

```python
top
```

Revenue per employee is a financial metric used to measure a company's efficiency in generating revenue relative to the number of employees. I will analyze Revenue per employee. if a company's revenue per employee is $200K USD, it means that on average, each employee contributes $200,000 to the company's total revenue.

```python
rev_bol=top['Revenue per employee ($K USD)']>850
```

 Sort values by 'Revenue per employee ($K USD)' in descending order. 
 This sorts the DataFrame based on the revenue generated per employee, 
 with the companies with the highest revenue per employee appearing first

```python
top_rev=top[rev_bol].sort_values('Revenue per employee ($K USD)',ascending=False)
top_rev
```

  Sort values by 'Number of Employees' in ascending order. 
 This sorts the DataFrame based on the number of employees in each company, 
#with companies having the fewest employees appearing first.

```python
least_emp=top_rev.sort_values('Employees')
least_emp
```

Now data_sorted DataFrame contains the sorted data with companies, 
 having the highest revenue per employee and fewest employees at the top.
