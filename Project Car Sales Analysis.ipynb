In this project, I will analyze car data from Germany to determine the most popular car based on price and mileage, while initially addressing any data bugs.

```python
import pandas as pd
```

```python
cars= pd.read_csv("cars.csv",encoding = 'latin-1') #bytes in the CSV file and convert them into text. In this case, 'latin-1' is the encoding scheme being used
```

```python
cars
```

```python
dir(cars['yearOfRegistration'])
```

```python
cars['yearOfRegistration'].value_counts().sort_index() #found bugs in data as there was no concept of cars before the 1800s
```

```python
restricted_years=cars['yearOfRegistration'].between(1800,2030) # made a boolean series and set range
```

```python
 normal_restriction_years=cars[restricted_years] #used boolean indexing to store True valued data in new variable 
```

```python
normal_restriction_years['yearOfRegistration'].value_counts().sort_index()
```

```python
normal_restriction_years
```

```python
normal_restriction_years['price'].value_counts().sort_index()# same as before found bugs as $0 cars 
```

```python
normal_restriction_years['price']=normal_restriction_years['price'].str.replace('$','').str.replace(',','').astype(int)
```

```python
normal_price_bol=normal_restriction_years['price'].between(1,350000)
```

```python
normal_price =normal_restriction_years[normal_price_bol]
```

```python
normal_price['price'].value_counts().sort_index()
```

Now data is cleaned and bugs are solved now to analysis.

```python
selling_prob =normal_price['brand'].value_counts(normalize=True)#value_counts a series method
selling_prob # brand col values are counted as indexes are repeated
```

```python
selling_bol = selling_prob >0.05#A boolean series condition is set
selling_prob[selling_bol]
```

```python
top_brands = selling_prob[selling_bol]
```

```python
top_brands
```

```python
brand_ind =top_brands.index
```

```python
brand_bol = normal_price["brand"].isin(brand_ind)#checked if brand data had the indexes from above
brand_bol
```

```python
top_selling_brands = normal_price[brand_bol].copy()
top_selling_brands
```

```python
top_selling_brands['odometer']
```

```python
top_selling_brands['odometer']=top_selling_brands['odometer'].str.replace(',','').str.replace('km','').astype(int)
```

```python
top_selling_brands['odometer'].value_counts().sort_index()
```

```python
top_selling_brands.rename({'price':'price ($)','odometer':'odometer (km)'}, axis=1,inplace=True)
```

```python
top_selling_brands #changed the name of columns 
```

```python
top_selling_brands.info()
```

```python
groupping=top_selling_brands.groupby("brand")
groupping 
```

```python
brand_mean = groupping[['price ($)','odometer (km)']].mean()
brand_mean
```

```python
brand_mean.sort_values('price ($)',ascending=False).sort_values('odometer (km)',ascending=False)
```
